```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
LINK_PREDICTION_RESULTS_DIR='/Users/ashfaq/Documents/alibaba_gnn_results'
```

```python
results_path = f'{LINK_PREDICTION_RESULTS_DIR}/results/_alibaba.pkl'
```

```python
with open(results_path, 'rb') as f:
    link_prediction_results = pickle.load(f)
```

```python
link_prediction_results
```

```python
# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot train losses
axes[0].plot(link_prediction_results['train_losses'], marker='o')
axes[0].set_title('Training Losses')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE')

# Plot validation APS
axes[1].plot(link_prediction_results['val_aps'], marker='o', color='orange')
axes[1].set_title('Validation APS')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('APS')

# Adjust layout
plt.tight_layout()
plt.savefig(f'../plots/metrics/link_prediction_metrics.png')
plt.show()
```
