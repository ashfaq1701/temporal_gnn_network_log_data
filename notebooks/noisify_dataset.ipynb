```python
from dotenv import load_dotenv
from src.utils import get_test_workloads
import matplotlib.pyplot as plt

load_dotenv()
```

```python
MICROSERVICE_ID = 25
```

```python
test_workloads = get_test_workloads(MICROSERVICE_ID)
```

```python
plt.plot(test_workloads)
```

```python
import numpy as np
import matplotlib.pyplot as plt

test_workloads = np.array(test_workloads)

iterations = 10

max_value = np.max(test_workloads)
random_noise = np.random.normal(0, 1, len(test_workloads))
shifted_noise = random_noise - np.min(random_noise)
normalized_noise = shifted_noise / np.max(shifted_noise)
scaled_noise = normalized_noise * max_value

modulation_factor_start = 1
modulation_factor_end = 0

modulation_decrease = (modulation_factor_start - modulation_factor_end) / (iterations - 1)

fig, axes = plt.subplots(iterations, 1, figsize=(10, iterations * 3))

for i in range(iterations):
    current_modulation_factor = modulation_factor_start - i * modulation_decrease

    noisy_workloads = (1.0 - current_modulation_factor) * scaled_noise + current_modulation_factor * test_workloads

    noisy_workloads[noisy_workloads < 0] = 0

    axes[i].plot(noisy_workloads, label=f'Iteration {i + 1} (Modulation Factor = {current_modulation_factor:.2f})',
                 alpha=0.7)
    axes[i].set_title(f"Iteration {i + 1} (Noise Decreases, Modulation Factor = {current_modulation_factor:.2f})")
    axes[i].legend()
    axes[i].set_xlabel("Index")
    axes[i].set_ylabel("Workload Value")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

test_workloads = np.array(test_workloads)

iterations = 10

max_value = np.max(test_workloads)
random_noise = np.random.normal(0, 1, len(test_workloads))
shifted_noise = random_noise - np.min(random_noise)
normalized_noise = shifted_noise / np.max(shifted_noise)
scaled_noise = normalized_noise * max_value

modulation_factor_start = 0
modulation_factor_end = 1

modulation_decrease = (modulation_factor_end - modulation_factor_start) / (iterations - 1)

fig, axes = plt.subplots(iterations, 1, figsize=(10, iterations * 3))

for i in range(0, iterations):
    current_modulation_factor = modulation_factor_start + i * modulation_decrease
    
    noisy_workloads = (1.0 - current_modulation_factor) * scaled_noise + current_modulation_factor * test_workloads
    
    noisy_workloads[noisy_workloads < 0] = 0
    
    axes[i].plot(noisy_workloads, label=f'Iteration {i+1} (Modulation Factor = {current_modulation_factor:.2f})', alpha=0.7)
    axes[i].set_title(f"Iteration {i+1} (Noise Decreases, Modulation Factor = {current_modulation_factor:.2f})")
    axes[i].legend()
    axes[i].set_xlabel("Index")
    axes[i].set_ylabel("Workload Value")

plt.tight_layout()
plt.show()
```
